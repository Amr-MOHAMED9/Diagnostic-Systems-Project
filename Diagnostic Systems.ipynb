# Defining homogenity

```python
#defining homogenity as the sum of all size/distance measures 
colnames = ['area','concavity','texture','perimeter','radius','concave points','smoothness','symmetry']
indexes=['_0','_1','_2']
```

```python
#get features with highest mean
features = []
for k in colnames:
    mean = 0
    varName = ''
    for j in indexes:
        if data[k+j].mean() > mean:
            mean = data[k+j].mean()
            varName = k+j
    features.append(varName)
```

```python
features.insert(0,'id')
features.insert(1,'malignant')
```

```python
features
```

```python
featuresSelected = pd.DataFrame(data[features], columns=features)
```

```python
featuresSelected.head()
```

```python
#calculating homogenity
featuresSelected['homogenity'] = featuresSelected.area_2+featuresSelected.perimeter_2+featuresSelected.radius_2 + featuresSelected.smoothness_2+featuresSelected.symmetry_2


#scaling the data
for i in featuresSelected.columns[2:]:
    featuresSelected[i] = featuresSelected[i]/ max(featuresSelected[i])
```

```python
featuresSelected.head()
```

```python
# ANOVA one-way test to validate the hypothesis of the difference
#in means of the 2 groups malignant/enign of each feature
for i in features[2:7]+features[8:]: 
    print('\n'+i+  ':\n')
    mod = ols('malignant ~'+i,
                    data=data).fit()

    aov_table = sm.stats.anova_lm(mod, typ=2)
    print(aov_table)
```

# rule based

```python
def perf_measure(y_actual, y_hat):
    TP = 0
    FP = 0
    TN = 0
    FN = 0

    for i in range(len(y_hat)): 
        if y_actual[i]==y_hat[i]==1:
           TP += 1
        if y_hat[i]==1 and y_actual[i]!=y_hat[i]:
           FP += 1
        if y_actual[i]==y_hat[i]==0:
           TN += 1
        if y_hat[i]==0 and y_actual[i]!=y_hat[i]:
           FN += 1
    TPR = 0
    TNR = 0
    TPR = round(TP/(TP+FN),2)
    TNR = round(TN/ (TN+FP),2)
    print('True positives: '+str(TP))
    print('False positives: '+str(FP))
    print('True negatives: '+str(TN))
    print('False negatives: '+str(FN))
    print('Sensitivity: '+str(TPR))
    print('Specifity: '+ str(TNR))
```

```python
# rule based model with threshold = feature mean + feature std
pred_list=[]
for i in featuresSelected.index:
    if featuresSelected.area_2[i] >  (featuresSelected.area_2.mean()+featuresSelected.area_2.std()):
        pred_list.append(1)
        
    elif featuresSelected['concave points_2'][i] >=  (featuresSelected['concave points_2'].mean()+featuresSelected['concave points_2'].std()):
        pred_list.append(1)
        
    elif featuresSelected.texture_2[i] >=  (featuresSelected.texture_2.mean()+featuresSelected.texture_2.std()):
        
        pred_list.append(1)
        
    elif featuresSelected.homogenity[i] >=  (featuresSelected.homogenity.mean()+featuresSelected.homogenity.std()):
        pred_list.append(1)
        
    else:
        pred_list.append(0)
```

```python
accuracy = sum([1 if a == b else 0 for a,b in zip(featuresSelected.malignant, pred_list)])/featuresSelected.malignant.shape[0]
print("The test-accuracy is {:.5f}".format(accuracy))
```

```python
perf_measure(featuresSelected.malignant, pred_list)
```

```python
# rule based model with threshold = feature's malignant cells group mean
pred_list=[]
for i in featuresSelected.index:
    if featuresSelected.area_2[i] > featuresSelected[featuresSelected.malignant==1].area_2.mean():
        pred_list.append(1)
        
    elif featuresSelected['concave points_2'][i] >=  (featuresSelected[featuresSelected.malignant==1]['concave points_2'].mean()):
        pred_list.append(1)
        
    elif featuresSelected.texture_2[i] >=  (featuresSelected[featuresSelected.malignant==1].texture_2.mean()):
        
        pred_list.append(1)
        
    elif featuresSelected.homogenity[i] >=  (featuresSelected.homogenity[featuresSelected.malignant==1].mean()):
        pred_list.append(1)
        
    else:
        pred_list.append(0)
```

```python
accuracy = sum([1 if a == b else 0 for a,b in zip(featuresSelected.malignant, pred_list)])/featuresSelected.malignant.shape[0]
print("The test-accuracy is {:.5f}".format(accuracy))
```
